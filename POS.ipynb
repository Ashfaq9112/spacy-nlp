```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Tyrion lannister is smart and sketchy and he is so badass")
```

```python
for d in doc:
    print(d,"|",d.pos_)
```

```python
doc = nlp("Wow john made 256 million $ on the very first day")
```

```python
for d in doc:
    print(d,"|",d.pos_,"|",spacy.explain(d.pos_),"|",d.tag_,"|",spacy.explain(d.tag_))
```

```python
d = nlp("he runs for the train")
print(d[1].text,d[1].tag_,spacy.explain(d[1].tag_))
```

```python
text =  """Microsoft Corp. today announced the following results for the quarter ended December 31, 2021, as compared to the corresponding period of last fiscal year:

·         Revenue was $51.7 billion and increased 20%
·         Operating income was $22.2 billion and increased 24%
·         Net income was $18.8 billion and increased 21%
·         Diluted earnings per share was $2.48 and increased 22%
“Digital technology is the most malleable resource at the world’s disposal to overcome constraints and reimagine everyday work and life,” said Satya Nadella, chairman and chief executive officer of Microsoft. “As tech as a percentage of global GDP continues to increase, we are innovating and investing across diverse and growing markets, with a common underlying technology stack and an operating model that reinforces a common strategy, culture, and sense of purpose.”
“Solid commercial execution, represented by strong bookings growth driven by long-term Azure commitments, increased Microsoft Cloud revenue""" 
```

```python
doc = nlp(text)
```

```python
for d in doc:
    print(d,"|",d.pos_)
```

```python
#now as you see there are extra other words punctuation and stuffs.... we have to remove those things
```

```python
for d in doc:
    if d.pos_ in ['SPACE','PUNCT','X']:
        print(d,"|",d.pos_ )
```

```python
#filtering tokens
filtered_tokens = []
for d in doc:
    if d.pos_ not in ['SPACE','PUNCT','X']:
        filtered_tokens.append(d)
```

```python
filtered_tokens
```

```python
#now countings verbs and nounns using countby
count = doc.count_by(spacy.attrs.POS)
count
```

```python
for k, v in count.items():
    print(doc.vocab[k].text,"|",v)
```
